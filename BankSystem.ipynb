## Account
* This is a class which make a new bank account with there unique account number, bank holder name and some initial balance and It works for deposit money in an account, withdraw money if bank balance is sufficient for this deducts  and get information for there account balance.


```python
class Account:
    def __init__(self, account_number, owner_name, balance):# we make new account holder with these attributes
        self.account_number= account_number
        self.owner_name= owner_name
        self.balance= balance # that is the initial deposit

    def deposit(self, amount):
        self.balance+= amount
        print(f"You deposit {amount}.New balance {self.balance}")

    def withdraw(self, amount):
        if (self.balance> amount):
            self.balance-= amount
            print(f"You withdraw {amount}. New balance {self.balance}")
        else:
            print("Your account balance is not sufficient for deducts such amount of money")

    def get_balance(self):
        return self.balance
```

## Current Account
* Basically this is also an account which has an overdraft limit so that you couldn't cross your exceed amount above the overdraft limit.

```python
# here we make Saving account where account class is super class or
# SavingsAccount is inherits from the Account class
class SavingsAccount(Account):
    def __init__(self, account_number, owner_name, balance, interest_rate):
        super().__init__(account_number, owner_name, balance)
        self.interest_rate= interest_rate

    def add_interest(self):
        interest= self.balance* self.interest_rate
        self.balance+= interest
        print(f"Interest {interest} is added.Now the new balance:{self.balance}")

# Now we make Current Account class
# and also that current account class inherits from the Account class
# means it's also we a account but some extra attributes
class CurrentAccount(Account):
    def __init__(self, account_number, owner_name, balance, overdraft_limit):
        super().__init__(account_number, owner_name, balance)
        self.overdraft_limit= overdraft_limit

    def withdraw(self, amount):
        if (amount- self.balance> self.overdraft_limit):
            print(f"withdrawal exceeds overdraft limit")
        else:
            self.balance-= amount
            print(f"You withdraw {amount}.New balance is {self.balance}")
```

## Bank
* This class is the collection of the accounts and work for adding new account, removing an account and finding an account is present in the bank account list or not.

```python
class Bank:
    def __init__(self): # accounts are a list of all accounts
        self.accounts= []

    def add_account(self, account):
        self.accounts.append(account)
        print(f"Account {account.account_number} added")

    def remove_account(self, account_number):
        self.accounts= [account for account in self.accounts if account.account_number!= account_number]
        print(f"Account {account_number} removed.")

    def find_account(self, account_number):
        for account in self.accounts:
            if (account.account_number== account_number):
                print(f"Yes, there is an account whose account number is {account_number}.")
                return account
            else:
                print(f"There is no such account which account number is {account_number}")
                return None
```

* Some usages:

```python
# Examples Usages

# Create a bank
my_bank= Bank()

# add accounts
simple= Account(100, "ayush", 200)
savings= SavingsAccount(101, "aditya", 5000, 0.02) # so that is a saving account in initial balance=5000 and interest rate 2%
current= CurrentAccount(102, "arpita", 500, overdraft_limit=300) # so that is a current account in which the initial balace is 500

my_bank.add_account(simple)
my_bank.add_account(savings)
my_bank.add_account(current)

# perform transactions
print(f"For simple account.")
simple.deposit(100)
simple.withdraw(200)
simple.get_balance

# second for savings account
print(f"For saving account.")
savings.deposit(1000)
savings.withdraw(2000)
savings.add_interest()

print(f"For current account.")
current.deposit(100)
current.withdraw(1000) # if we try to cross overdraft limit
# if we do not want to cross overdraft limit eg
current.withdraw(800)
```
